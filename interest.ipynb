```python
def calculer_interet():
    try:
        montant = float(input("Montant total à placer (en €) : "))
        annees = int(input("Nombre d'années d'investissement : "))
        taux = float(input("Taux d'intérêt annuel (ex: 0.02 pour 2%) : "))
        
        if montant < 0 or annees < 0 or taux < 0:
            print("Erreur : Les valeurs ne peuvent pas être négatives.")
            return
        
        if taux > 1:
            print("Erreur : Le taux doit être un nombre décimal (ex: 0.02 pour 2%), pas un pourcentage.")
            return
        
        montant_final = montant * ((1 + taux) ** annees)
        
        print(f"\n Après {annees} ans, vous aurez : {montant_final:.2f} €")
        
    except ValueError:
        print("Erreur : Vous devez saisir des nombres valides.")
```

```python
calculer_interet()
```
